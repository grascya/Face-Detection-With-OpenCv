```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
turing =  cv2.imread('turing.jpg',0)
tesla =cv2.imread('tesla.jpg',0)
solvay = cv2.imread('solvay.jpg',0)
```

```python
plt.imshow(solvay, cmap='gray')
```

```python
cascade = cv2.CascadeClassifier('haarcascades/haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    
    face_img = img.copy()
  
    face_rectangle = cascade.detectMultiScale(face_img) 
    
    for (x,y,w,h) in face_rectangle: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,255,255), 10) 
        
    return face_img
```

```python
res = detect_face(turing)
plt.imshow(res, cmap='gray')
```

```python
def adjusted_detect_face(img):
    
    face_img = img.copy()
  
    face_rectangle = cascade.detectMultiScale(face_img, scaleFactor=1.2, minNeighbors=5,minSize=(30,30)) 
    
    for (x,y,w,h) in face_rectangle: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,255,255), 10) 
        
    return face_img
```

```python
result = adjusted_detect_face(tesla)
plt.imshow(result,cmap='gray')
```

```python
eye_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_eye.xml')
```

```python
def detect_eyes(img):
    
    face_img = img.copy()
  
    eyes = eye_cascade.detectMultiScale(face_img,scaleFactor=1.6, minNeighbors=1,minSize=(30,30)) 
    
    
    for (x,y,w,h) in eyes: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,255,255), 10) 
        
    return face_img
```

```python
result = detect_eyes(turing)
plt.imshow(result,cmap='gray')
```

```python
cap = cv2.VideoCapture(0) 

while True: 
    
    ret, frame = cap.read(0) 
     
    frame = detect_face(frame)
 
    cv2.imshow('Video Face Detection', frame) 
 
    c = cv2.waitKey(1) 
    if c == ord('q'): 
        break 
        
cap.release() 
cv2.destroyAllWindows()
```
